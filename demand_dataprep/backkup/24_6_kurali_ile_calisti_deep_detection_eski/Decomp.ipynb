```python
import pandas as pd
from statsmodels.tsa.seasonal import STL
import warnings
from sklearn.linear_model import LinearRegression
import numpy as np
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
warnings.filterwarnings("ignore")
```

```python
df = pd.read_excel('../data/_pasifik_sales_portfoy.xlsx')
df.head()
```

```python
df_migros = df[df['grup_adi'] == 'MİGROS']
df_migros_bisk = df_migros[df_migros['kategori_adi'] == 'BİSKÜVİ']
df_migros_bisk.head()
```

```python
df_migros_bisk.date = pd.to_datetime(df_migros_bisk.date)
```

```python
df_mig_bis = df_migros_bisk.groupby(['yil','ay','date','kategori_adi']).agg({'adet':'sum'})
```

```python
df_mig_bis
```

```python
df_mig_bis.reset_index(inplace=True)
```

```python
df_mig_bis
```

```python
df_ts = df_mig_bis[['adet','date']]
df_ts.set_index('date',inplace=True)
```

```python
df_ts
```

```python
result = STL(df_ts).fit()
df_mig_bis['season'] = list(result.seasonal)
df_mig_bis['trend']  = list(result.trend)
df_mig_bis['residual']  = list(result.resid)
```
